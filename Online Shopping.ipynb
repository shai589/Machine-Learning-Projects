```python
# Loading the dataset
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report


# Update with your actual file path
file_path = r"/Users/tasnimshairy/Desktop/online_shoppers_intention.csv"

# Load the CSV file
df = pd.read_csv(file_path)
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

# 1. Perform one-hot encoding for 'Month' and 'VisitorType'
df_encoded = pd.get_dummies(df, columns=['Month', 'VisitorType'])

# 2. Convert only the new one-hot encoded columns to integers
# Identify the one-hot encoded columns
encoded_columns = [col for col in df_encoded.columns if 'Month_' in col or 'VisitorType_' in col]

# Convert these columns to integers
df_encoded[encoded_columns] = df_encoded[encoded_columns].astype(int)

# 3. Convert 'Revenue and Weekend' column to binary (1 for True, 0 for False)
df_encoded['Revenue'] = df_encoded['Revenue'].astype(int)
df_encoded['Weekend'] = df_encoded['Weekend'].astype(int)
# Display the first few rows of the preprocessed dataset
print(df_encoded.head())
```

```python
# Model Building Imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import f1_score, recall_score, roc_auc_score
```

```python
# Identify target (Y) and predictor (X) variables
# Features (X): All columns except Revenue
X = df_encoded.drop(columns=['Revenue'])

# Target (y): Revenue column
y = df_encoded['Revenue']
```

```python
#Split data into training and testing sets

# Train test split
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=20)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.25, stratify=y_train_val, random_state=20)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Organize your train, validation, and test splits into a dictionary
data_model = {
    "Train": (X_train, y_train),
    "Validation": (X_val, y_val),
    "Test": (X_test, y_test)
}

# Loop through each dataset in the dictionary
for name, (X_data, y_data) in data_model.items():
    # Create a temporary DataFrame for the features
    df_temp = pd.DataFrame(X_data.copy(), columns=X.columns)  # Use X's columns for consistency
    # Add the target column to the temporary DataFrame
    df_temp['Revenue'] = y_data.values
    # Print summary information
    print(f'\n{name} Dataset:')
    print(f'Number of rows: {df_temp.shape[0]}')
    print(f'Number of columns: {df_temp.shape[1]}')
    print('First 5 rows:')
    print(df_temp.head())  # Display the first 5 rows
    print('-' * 80)
```
